<h4>Plot</h4>
plot with X and Y can be used to plot the equation and graph distribution

```python
#x2=4*ay
#y=x2/(4a)

def parabola_plot(x,a=1):
    y=[i**2/(4*a) for i in x]
    return y
    
X=[-3,-2,-1,0,1,2,3]

Y=parabola_plot(X)
plt.plot(X,Y)
display(X,Y)
```

<h3>Scatter plot</h3>

```python
import random 
physics_marks=[random.randrange(1,100) for i in range(1,100)]
mathematics_marks=[random.randrange(1,100) for i in range(1,100)]
plt.scatter(physics_marks,mathematics_marks)
plt.xlabel('Physics')
plt.ylabel('Mathematics')
```
