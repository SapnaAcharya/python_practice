Question 1: What was the best month for sales? How much was earned that month?

```python
results = all_data.groupby('Month').sum()
results
# It only sums up the columns with numeric datatype!
```

Let's plot this!

```python
import matplotlib.pyplot as plt

months = range(1,13)
plt.figure(figsize=(9,4))
plt.bar(months,results['Sales'])
# plt.bar (results.index, results['Sales]) <- can we use this too
plt.xticks(months)
plt.ylabel('Total Sales in a month in USD ($)')
plt.grid(axis='y')
plt.show()
```

```python
print(f"Thus, the best month in sales was Month {results.idxmax()['Sales']} with a sale of " + "${:,.2f}".format(results.max()['Sales']))
```

So, April being the best month in sales followed by February, July and march can be explained by the festivities that are in these months. Generally in April, shopping spends peak around .

Question 2: What time should we display advertisements to maximize likelihood of customers buying products?



First converting Order Date column from str to datetime object.

```python
all_data['Order Date'] = pd.to_datetime(all_data['Order Date'])
all_data.head()
```

```python
all_data['Hour'] = all_data['Order Date'].dt.hour
all_data['Minute'] = all_data['Order Date'].dt.minute
all_data.head()
```

```python
results = all_data.groupby(['Hour']).count()
results
```

```python
hours = results.index
plt.plot(hours, results['Order ID'])
plt.xticks(hours)
plt.xlabel('Hour of the Day')
plt.ylabel('Total Number of sales in an Hour')
plt.grid()
plt.show()
```

From the chart above, it is clearly evident that the peaks in shopping occur around 12pm (1200 hrs) and 7pm (1900 hrs) across the entire 10 US cities.
